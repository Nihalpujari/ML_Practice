```python
import nltk
from nltk.stem import PorterStemmer, WordNetLemmatizer,SnowballStemmer,LancasterStemmer
import spacy

nlpd=spacy.load('de_core_news_sm')
nlpe=spacy.load('en_core_web_sm')

text="""
1. OMGGG 😭😭 my flight from FRA → DEL got delayed AGAIN!!! #Lufthansa @airporthelp
2. The baggage tracking system isn't working properly after the latest update :(
3. GREAT SERVICE!!! waited ONLY 5 hours for support 👏👏 #sarcasm
4. Sehr schlechter Kundenservice!!! Niemand antwortet auf meine Emails.
5. Flight AI-202 cancelled 😡 Refund still pending after 3 weeks!!!
6. Visit www.airlinehelp.com/refund to check your refund status.
7. I can’t login to the app since update v2.4.1 pls fix ASAP!!!!
8. 😠😡
9. Seats were fine but food was meh...
10. SUPPORT SUPPORT SUPPORT!!!!!!!
"""
tokens=nltk.word_tokenize(text)

stemmer=PorterStemmer()
lemmitazer=WordNetLemmatizer()
snowball=SnowballStemmer("english")
lancaster=LancasterStemmer()

words=['tracking']
snowballed=[snowball.stem(word) for word in words]
lancastered=[lancaster.stem(word) for word in words]
stemmed=[stemmer.stem(word) for word in words]
lemmitazed=[lemmitazer.lemmatize(word) for word in words]

print("here is snowballed words",snowballed)
print("here is lancastered words",lancastered)
print("here is stemmed words",stemmed)
print("here is lemmatized words",lemmitazed)
```
