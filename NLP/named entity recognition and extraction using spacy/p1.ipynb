```python
import spacy
nlp = spacy.load('en_core_web_sm')
text="my name is Nihal i study in SRH university from Sept 2025."
doc = nlp(text)
```

```python
for ent in doc.ents:
    print(ent.text,ent.label_)
```

```python
for ent in doc.ents:
    print(f"{ent.text:<25},{ent.label_:<20}{spacy.explain(ent.label)}")
```

```python
from spacy import displacy

html= displacy.render(doc, style="ent")
display(html)
```
