```python
import re
text=""" Ich bin nihal, Ich bin 20 Jahre alt, ich bin Studentin, ich wohne in Heilberg, meine Telefonnummer ist 15510733665"""
```

```python
numbers=re.findall(r'\d+', text)    
print(numbers)
```

```python
tex1=""" contact me at nihal.pujari@srh.de
or support@gmail.com"""

emails=re.findall(r'[\w\.]+@[\w\.]+', tex1)
print(emails)
```

```python
text2="hello!! how are you??"
punctuation=re.sub(r'[^\w\s]','', text2)
print(punctuation)
```
