```python
import spacy
nlp = spacy.load("en_core_web_lg")

text = """
The large pre-trained language  from OpenAI has changed natural language processing.
Many research  at leading universities are now studying its emergent capabilities.
The  of France.
New York .
the  that OpenAI built
"""
 
doc = nlp(text)
for sent in doc.sents:
    chunks = list(sent.noun_chunks)
    if chunks:
        print(f"{str(sent).strip():<70} main noun = {chunks[0].root.text}")
```
