```python
import gensim.downloader as api
model=api.load('glove-wiki-gigaword-50')
print(type(model))
```

```python
print(model['movie'])
```

```python
print(model.similarity('movie','film'))
```

```python
print(model.most_similar('movie',topn=10))
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
words=['movie','film','bus','car','train','actor','dog','cat','king','queen']
```

```python
words_vectors=[model[word] for word in words]
pca=PCA(n_components=2)
result=pca.fit_transform(words_vectors)
plt.scatter(result[:,0],result[:,1])
for i,word in enumerate(words):
    plt.annotate(word,(result[i,0],result[i,1]))
plt.title('PCA of Word Vectors')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.xticks(rotation=90)
plt.yticks(rotation=0)
plt.show()
```
