```python
from transformers import pipeline
from sklearn.metrics import accuracy_score

# Both PyTorch models - work locally without TensorFlow
pipe_a = pipeline(task="sentiment-analysis", model="distilbert-base-uncased-finetuned-sst-2-english", framework="pt")
pipe_b = pipeline(task="sentiment-analysis", model="cardiffnlp/twitter-roberta-base-sentiment-latest", framework="pt")

texts = ["I love this!", "This is terrible", "Great product", "Worst experience ever"]
true_labels = ["POSITIVE", "NEGATIVE", "POSITIVE", "NEGATIVE"]

preds_a = [res["label"] for res in pipe_a(texts)]
preds_b = [res["label"] for res in pipe_b(texts)]

acc_a = accuracy_score(preds_a, true_labels)
acc_b = accuracy_score(preds_b, true_labels)

print(f"Accuracy - Model A: {acc_a:.2f}")
print(f"Accuracy - Model B: {acc_b:.2f}")
```
