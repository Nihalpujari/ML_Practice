```python
import pandas as pd
import numpy as np
import random

# Load your dataset (replace 'your_dataset.csv' with your file name)
df = pd.read_csv('diabetes_clean.csv')

# Lists of common first and last names
df.head()
```

```python
first_names = [
    'John', 'Emma', 'Michael', 'Sarah', 'David', 'Lisa', 'James', 'Anna', 
    'Robert', 'Maria', 'William', 'Jennifer', 'Chris', 'Amanda', 'Mark', 
    'Jessica', 'Paul', 'Laura', 'Kevin', 'Michelle', 'Brian', 'Nicole',
    'Ryan', 'Ashley', 'Jason', 'Stephanie', 'Eric', 'Melissa', 'Justin',
    'Heather', 'Matthew', 'Amber', 'Andrew', 'Rachel', 'Daniel', 'Angela'
]

last_names = [
    'Smith', 'Johnson', 'Williams', 'Brown', 'Jones', 'Garcia', 'Miller',
    'Davis', 'Rodriguez', 'Martinez', 'Hernandez', 'Lopez', 'Gonzalez',
    'Wilson', 'Anderson', 'Thomas', 'Taylor', 'Moore', 'Jackson', 'Martin',
    'Lee', 'Perez', 'Thompson', 'White', 'Harris', 'Sanchez', 'Clark',
    'Ramirez', 'Lewis', 'Robinson', 'Walker', 'Young', 'Allen', 'King'
]

# Generate EXACTLY 768 unique full names
random.seed(42)  # For reproducibility
full_names = []
used_names = set()

while len(full_names) < 768:
    first = random.choice(first_names)
    last = random.choice(last_names)
    full_name = f"{first} {last}"
    
    if full_name not in used_names:
        used_names.add(full_name)
        full_names.append(full_name)

# Add names as a new column
df['Person_Name'] = full_names[:len(df)]  # Ensure exact match with your 768 rows

# Save the updated dataset
df.to_csv('dataset_with_names.csv', index=False)

print("✅ Added 768 unique names to your dataset!")
print("First 10 names:")
print(df['Person_Name'].head(10).tolist())
print(f"\nDataset shape: {df.shape}")
```
